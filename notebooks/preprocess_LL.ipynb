```python
import numpy as np
import pandas as pd
import csv
```

```python
#csv_file_path = 
train_df = pd.read_csv('../raw_data/sales_train_validation.csv')
prices_df = pd.read_csv('../raw_data/sell_prices.csv')
calendar_df = pd.read_csv('../raw_data/calendar.csv')
```

```python
top_200_products = pd.read_csv("../raw_data/top_200_product_names.csv")
```

```python
top_200_products = top_200_products["Product Name"].tolist()
```

```python
train_df.head()
```

```python
prices_df.head()
```

```python
calendar_df.head()
```

```python
train_df.shape
```

```python
filtered_train_df = train_df[train_df['id'].isin(top_200_products)]
```

```python
filtered_train_df
```

```python
train_df_sample = filtered_train_df
```

```python
train_df_sample = train_df_sample.melt(id_vars=['id','item_id','dept_id','cat_id','store_id','state_id'])
```

```python
train_df_sample.rename(columns={'variable': 'd', 'value':'sales'},inplace=True)
```

```python
merge_df = train_df_sample.merge(calendar_df,on='d',how='left')
merge_df.head()
```

```python
merge_df = merge_df.merge(prices_df,on=['store_id', 'item_id','wm_yr_wk'],how='left')
merge_df
```
