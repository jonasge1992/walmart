```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../raw_data/merge_df_scaled.csv")
```

```python
df
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df.set_index("date",inplace=True)
```

```python
df
```

```python
item_list = df["id"].unique()
```

```python
from sklearn.metrics import mean_absolute_error

mae_values = []

for item in item_list:
    item_df = df[df["id"] == item]["sales"]
    item_df_train = item_df.iloc[:-28]
    forecast = item_df_train.iloc[-1]
    item_df_test = item_df.iloc[-28:]
    
    mae = mean_absolute_error(item_df_test, [forecast]*len(item_df_test))
    
    # Append MAE to the list of MAE values
    mae_values.append(mae)
```

```python
np.mean(mae_values)
```

```python
for item in item_list[:10]:
    item_df = df[df["id"] == item]["sales"].iloc[-365:]
    plt.figure(figsize=(10, 6))
    plt.plot(item_df.index, item_df.values, label='Actual Sales')
    plt.title(f"Time Series Plot for Item {item}")
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.legend()
    plt.show()
```
